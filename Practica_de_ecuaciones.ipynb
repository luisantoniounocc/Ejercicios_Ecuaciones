EJERCICIOS DE SISTEMAS DE ECUACIONES

Ejercicio nº 1.-
a)	Resuelve por sustitución:
 5x  2y  1
3x  3y  5

b)	Resuelve por reducción:
 2x  y  6
4x  3y  14


```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_sistema_con_nombres(A, b, titulo, etiquetas):
    sol = np.linalg.solve(A, b)
    x_sol, y_sol = sol

    x_vals = np.linspace(x_sol - 5, x_sol + 5, 400)
    y1 = (b[0] - A[0,0]*x_vals) / A[0,1]
    y2 = (b[1] - A[1,0]*x_vals) / A[1,1]

    plt.figure(figsize=(8,6))
    plt.plot(x_vals, y1, label=etiquetas[0])
    plt.plot(x_vals, y2, label=etiquetas[1])
    plt.plot(x_sol, y_sol, 'ro', label='Solución')
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
    return sol

# a) Sustitución
A_sust = np.array([[5, 2], [-3, 3]])
b_sust = np.array([1, 5])
sol_sust = graficar_sistema_con_nombres(A_sust, b_sust, "Sistema a) - Sustitución", ["5x + 2y = 1", "-3x + 3y = 5"])
print("a) Método de sustitución:")
print(f"x = {sol_sust[0]:.4f}, y = {sol_sust[1]:.4f}")

# b) Reducción
A_red = np.array([[2, 1], [4, 3]])
b_red = np.array([6, 14])
sol_red = graficar_sistema_con_nombres(A_red, b_red, "Sistema b) - Reducción", ["2x + y = 6", "4x + 3y = 14"])
print("\nb) Método de reducción:")
print(f"x = {sol_red[0]:.4f}, y = {sol_red[1]:.4f}")
```

Ejercicio nº 6.-

Resuelve cada uno de los siguientes sistemas:

a)	 x  2y  1
3x  y  10

b)	 x  2y  4
2x  4y  3


```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_sistema_estable(A, b, titulo, etiquetas):
    det = np.linalg.det(A)
    x_vals = np.linspace(-10, 10, 400)  # Rango fijo para todos los gráficos

    y1 = (b[0] - A[0,0]*x_vals) / A[0,1]
    y2 = (b[1] - A[1,0]*x_vals) / A[1,1]

    plt.figure(figsize=(8,6))
    plt.plot(x_vals, y1, label=etiquetas[0])
    plt.plot(x_vals, y2, label=etiquetas[1])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(titulo)
    plt.grid(True)
    plt.legend()
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    if abs(det) > 1e-12:
        sol = np.linalg.solve(A, b)
        plt.plot(sol[0], sol[1], 'ro', label='Solución')
        plt.legend()
        print(f"Solución: x = {sol[0]:.4f}, y = {sol[1]:.4f}")
    else:
        plt.text(0, 0, 'No hay solución única\n(matriz singular o sistema inconsistente)',
                 fontsize=12, color='red', ha='center')
        print("No hay solución única (matriz singular o sistema inconsistente)")

    plt.show()


# Sistema a)
A_a = np.array([[1, 2], [-3, 1]])
b_a = np.array([1, -10])
graficar_sistema_estable(A_a, b_a, "Sistema a)", ["x + 2y = 1", "-3x + y = -10"])

# Sistema b)
A_b = np.array([[-1, 2], [2, -4]])
b_b = np.array([4, 3])
graficar_sistema_estable(A_b, b_b, "Sistema b)", ["-x + 2y = 4", "2x - 4y = 3"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def resolver_y_graficar_sistema(A, b, titulo, etiquetas):
    """
    Resuelve y grafica un sistema de 2x2 de la forma Ax = b.
    A: Matriz de coeficientes.
    b: Vector de términos independientes.
    titulo: Título para el gráfico.
    etiquetas: Etiquetas para las ecuaciones en la leyenda.
    """
    print(f"\n--- {titulo} ---")

    # Intenta resolver el sistema de ecuaciones
    try:
        sol = np.linalg.solve(A, b)
        print(f"Solución: x = {sol[0]:.4f}, y = {sol[1]:.4f}")

        # Generar puntos para graficar las líneas
        x_vals = np.linspace(-10, 10, 400)
        y1 = (b[0] - A[0, 0] * x_vals) / A[0, 1]
        y2 = (b[1] - A[1, 0] * x_vals) / A[1, 1]

        # Graficar las líneas y la solución
        plt.figure(figsize=(8, 6))
        plt.plot(x_vals, y1, label=etiquetas[0])
        plt.plot(x_vals, y2, label=etiquetas[1])
        plt.plot(sol[0], sol[1], 'ro', markersize=10, label=f'Solución: ({sol[0]:.2f}, {sol[1]:.2f})')
        plt.title(titulo + ": Solución Única")

    # Si el sistema no tiene una solución única (determinante = 0)
    except np.linalg.LinAlgError:
        print("El sistema no tiene una solución única.")
        print("Geométricamente, las líneas son paralelas o la misma línea.")

        # Generar puntos para graficar las líneas
        x_vals = np.linspace(-10, 10, 400)
        y1 = (b[0] - A[0, 0] * x_vals) / A[0, 1]
        y2 = (b[1] - A[1, 0] * x_vals) / A[1, 1]

        # Graficar las líneas sin un punto de intersección
        plt.figure(figsize=(8, 6))
        plt.plot(x_vals, y1, label=etiquetas[0])
        plt.plot(x_vals, y2, label=etiquetas[1])
        plt.title(titulo + ": Sin Solución Única")

    # Configuración final del gráfico
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.legend()
    plt.show()

# --- EJECUCIÓN DE LOS EJERCICIOS ---

# a) Sistema de ecuaciones: 2x + 3y = 1, 3x + 2y = 4
A1 = np.array([[2, 3], [3, 2]])
b1 = np.array([1, 4])
etiquetas1 = ["2x + 3y = 1", "3x + 2y = 4"]
resolver_y_graficar_sistema(A1, b1, "Ejercicio 9 - a", etiquetas1)

# b) Sistema de ecuaciones: 4x - 3y = 5, -8x + 6y = 10
A2 = np.array([[4, -3], [-8, 6]])
b2 = np.array([5, 10])
etiquetas2 = ["4x - 3y = 5", "-8x + 6y = 10"]
resolver_y_graficar_sistema(A2, b2, "Ejercicio 9 - b", etiquetas2)
```

Ejercicio nº 21.- Averigua cuántas soluciones tiene el siguiente sistema de ecuaciones, representando las dos rectas en los mismos ejes:

	x  y  5
2x  2y  2


```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes y términos independientes
A = np.array([[-1, 1],
              [-2, 2]])
b = np.array([5, 2])

try:
    # Intentamos resolver el sistema
    sol = np.linalg.solve(A, b)
    solucion_unica = True
except np.linalg.LinAlgError:
    solucion_unica = False

# Valores de x para graficar
x_vals = np.linspace(-10, 10, 400)

# Ecuaciones en forma y = ...
y1 = 5 + x_vals       # -x + y = 5  =>  y = x + 5
y2 = (2 + 2*x_vals)/2 # -2x + 2y = 2 => 2y = 2 + 2x => y = x + 1

plt.figure(figsize=(8,6))
plt.plot(x_vals, y1, label=r'$-x + y = 5$')
plt.plot(x_vals, y2, label=r'$-2x + 2y = 2$')

if solucion_unica:
    plt.plot(sol[0], sol[1], 'ro', label=f'Solución: ({sol[0]:.2f}, {sol[1]:.2f})')

plt.title('Sistema de ecuaciones')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.xlim(-10, 10)
plt.ylim(-10, 15)
plt.show()

if solucion_unica:
    print(f"El sistema tiene una solución única: x = {sol[0]:.4f}, y = {sol[1]:.4f}")
else:
    print("El sistema no tiene solución única (infinitas soluciones o ninguna solución).")
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[-1, 1],
              [-2, 2]])
b = np.array([1, 2])

# Comprobar rango para determinar tipo de solución
rango_A = np.linalg.matrix_rank(A)
rango_Ab = np.linalg.matrix_rank(np.c_[A, b])

if rango_A == rango_Ab == 1:
    print("Las rectas son coincidentes. Infinitas soluciones.")
elif rango_A < rango_Ab:
    print("El sistema no tiene solución.")
else:
    sol = np.linalg.solve(A, b)
    print(f"Solución única: x = {sol[0]:.4f}, y = {sol[1]:.4f}")

x_vals = np.linspace(-5, 5, 400)
y1 = (b[0] - A[0, 0]*x_vals) / A[0, 1]
y2 = (b[1] - A[1, 0]*x_vals) / A[1, 1]

plt.figure(figsize=(8,6))
plt.plot(x_vals, y1, label=r'$-x + y = 1$')
plt.plot(x_vals, y2, label=r'$-2x + 2y = 2$')

if rango_A == rango_Ab == 1:
    plt.title("Rectas coincidentes - Infinitas soluciones")
elif rango_A < rango_Ab:
    plt.title("Rectas paralelas - Sin solución")
else:
    plt.title("Sistema con solución única")

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

Problema nº 1.-
Calcula un número sabiendo que la suma de sus dos cifras es 10; y que, si invertimos el orden de dichas cifras, el número obtenido es 36 unidades mayor que el inicial.


```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Planteamiento de las Ecuaciones ---
# Suma de las cifras: x + y = 10  ->  y = -x + 10
# Número invertido es 36 unidades mayor: 10y + x = 10x + y + 36 -> y = x + 4

# --- 2. Definición de funciones ---
def ecuacion1(x):
    return -x + 10

def ecuacion2(x):
    return x + 4

# --- 3. Rango de valores para x ---
x_vals = np.linspace(0, 10, 400)
y_vals1 = ecuacion1(x_vals)
y_vals2 = ecuacion2(x_vals)

# --- 4. Calculo de la intersección ---
interseccion_x = 3
interseccion_y = 7

# --- 5. Gráfica ---
plt.figure(figsize=(10, 8))
plt.plot(x_vals, y_vals1, label='y = -x + 10 (Suma de cifras)', color='blue')
plt.plot(x_vals, y_vals2, label='y = x + 4 (Número invertido)', color='red')
plt.plot(interseccion_x, interseccion_y, 'o', color='green', markersize=10, label=f'Solución ({interseccion_x}, {interseccion_y})')
plt.annotate(f'({interseccion_x}, {interseccion_y})', (interseccion_x, interseccion_y), xytext=(0,10), textcoords='offset points', ha='center', color='green', fontsize=12)
plt.title('Solución del Problema nº 1')
plt.xlabel('Cifra de las decenas (x)')
plt.ylabel('Cifra de las unidades (y)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()

# --- 6. Resultados ---
print(f"Cifra decenas: {int(interseccion_x)}")
print(f"Cifra unidades: {int(interseccion_y)}")
print(f"Número original: {int(interseccion_x*10 + interseccion_y)}")
print(f"Número invertido: {int(interseccion_y*10 + interseccion_x)}")
```

Problema nº 2.-

En un triángulo rectángulo, uno de sus ángulos agudos es 12 mayor que el otro. ¿Cuánto miden sus tres ángulos?


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, Arc

# --- 1. Cálculo de los ángulos ---
# Sea x y x+12 las dos agudas; junto con 90 suman 180:
# x + (x + 12) + 90 = 180  ->  2x + 12 = 90  -> x = 39
angulo1 = 39  # grados (uno de los agudos)
angulo2 = angulo1 + 12  # 51 grados
angulo_recto = 90

# --- 2. Construcción geométrica del triángulo rectángulo ---
# Tomamos el vértice A en el origen (0,0) como el ángulo recto.
# Colocamos el cateto adyacente a lo largo del eje x con longitud 1 (B = (1,0)).
# El otro cateto tendrá longitud tan(angulo1) en y para que el ángulo en B sea angulo1.
a = 1.0
b = np.tan(np.deg2rad(angulo1))

A = (0.0, 0.0)      # vértice con 90°
B = (a, 0.0)        # vértice con angulo1 (39°)
C = (0.0, b)        # vértice con angulo2 (51°)

# --- 3. Gráfica 1: Triángulo con anotaciones de los ángulos ---
plt.figure(figsize=(8,8))
ax = plt.gca()

# Dibujar el triángulo como polígono
tri = Polygon([A, B, C], closed=True, facecolor='lightblue', edgecolor='black', alpha=0.7)
ax.add_patch(tri)

# Líneas de los lados
ax.plot([A[0], B[0]], [A[1], B[1]], color='black')
ax.plot([B[0], C[0]], [B[1], C[1]], color='black')
ax.plot([C[0], A[0]], [C[1], A[1]], color='black')

# Dibujar arcos para marcar los ángulos
arc_A = Arc(A, 0.4, 0.4, angle=0, theta1=0, theta2=90, linewidth=2, color='green')
ax.add_patch(arc_A)

arc_B = Arc(B, 0.4, 0.4, angle=180, theta1=0, theta2=-angulo1, linewidth=2, color='red')
ax.add_patch(arc_B)

arc_C = Arc(C, 0.4, 0.4, angle=270, theta1=0, theta2=angulo2, linewidth=2, color='orange')
ax.add_patch(arc_C)

# Anotaciones de texto con los valores de los ángulos
ax.text(A[0]+0.08, A[1]+0.08, f"{angulo_recto}°", fontsize=12, weight='bold', color='green')
ax.text(B[0]-0.15, B[1]+0.05, f"{angulo1}°", fontsize=12, weight='bold', color='red')
ax.text(C[0]+0.05, C[1]-0.05, f"{angulo2}°", fontsize=12, weight='bold', color='orange')

# Ajustes visuales
ax.set_aspect('equal', 'box')
pad = 0.5
ax.set_xlim(min(A[0], B[0], C[0]) - pad, max(A[0], B[0], C[0]) + pad)
ax.set_ylim(min(A[1], B[1], C[1]) - pad, max(A[1], B[1], C[1]) + pad)
ax.set_title('Triángulo rectángulo: visualización geométrica de los ángulos')
ax.axis('off')  # se ocultan ejes para una apariencia más clara
plt.show()

# --- 4. Gráfica 2: mostrando los tres ángulos ---
plt.figure(figsize=(8,5))
labels = ['Ángulo recto', 'Ángulo agudo 1', 'Ángulo agudo 2']
values = [angulo_recto, angulo1, angulo2]
plt.bar(labels, values, color=['green','red','orange'])
plt.ylabel('Medida (grados)')
plt.title('Comparación de las medidas de los ángulos del triángulo')
plt.ylim(0, 100)
for i, v in enumerate(values):
    plt.text(i, v + 2, str(v) + "°", ha='center', fontsize=11, weight='bold')

plt.show()

# --- 5. Resultado en texto ---
print("Medidas de los ángulos del triángulo:")
print(f"Ángulo recto: {angulo_recto}°")
print(f"Ángulo agudo 1: {angulo1}°")
print(f"Ángulo agudo 2: {angulo2}°")
```

```python
import matplotlib.pyplot as plt

# Datos
x = 15  # litros del primer líquido
y = 25  # litros del segundo

print(f"Litros del primer líquido (0,94 €/l): {x} l")
print(f"Litros del segundo líquido (0,86 €/l): {y} l")

# Gráfico: pastel para visualizar proporción
plt.figure(figsize=(5,5))
labels = [f'Primero (0,94 €/l) - {x} l', f'Segundo (0,86 €/l) - {y} l']
sizes = [x, y]
colors = ['skyblue','lightgreen']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Composición de la mezcla (40 litros)')
plt.axis('equal')
plt.show()
```

Problema nº 13.-
El doble de un número más la mitad de otro suman 7; y, si sumamos 7 al primero de ellos, obtenemos el quíntuplo del otro. Plantea un sistema de ecuaciones y resuélvelo para hallar dichos números.


```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Solución del sistema ---
x = 3  # primer número
y = 2  # segundo número

# --- Estadísticas básicas ---
total = x + y
promedio = (x + y) / 2
porcentajes = [x/total*100, y/total*100]

print("=== RESULTADOS ===")
print(f"Primer número: {x}")
print(f"Segundo número: {y}")
print(f"Suma total: {total}")
print(f"Promedio: {promedio}")
print(f"Porcentajes: Primer número {porcentajes[0]:.1f}%, Segundo número {porcentajes[1]:.1f}%")

# --- Tabla con pandas ---
data = {
    'Número': ['Primer número', 'Segundo número'],
    'Valor': [x, y],
    'Porcentaje %': porcentajes
}
df = pd.DataFrame(data)
display(df)

# --- Gráfico ---
plt.figure(figsize=(6,4))
bars = plt.bar(df['Número'], df['Valor'], color=['skyblue','lightgreen'])
plt.title('Solución del Problema nº 13')
plt.ylabel('Valor')
for i,v in enumerate(df['Valor']):
    plt.text(i, v+0.1, f"{v} ({porcentajes[i]:.1f}%)", ha='center', fontsize=11)
plt.ylim(0, max(df['Valor'])+1)
plt.show()
```

Problema nº 14.-

Dos de los ángulos de un triángulo suman 122. El tercero de sus ángulos excede en 4 grados al menor de los otros dos. ¿Cuánto miden los ángulos del triángulo?


```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, radians

# Cálculo de ángulos
m = 54
M = 122 - m
c = m + 4
angulos = [m, M, c]
print("Ángulos del triángulo (grados):", angulos)

# Construcción del triángulo (Ley de Senos)
A_ang, B_ang, C_ang = m, M, c
c_len = 1.0
k = c_len / sin(radians(C_ang))
a_len = k * sin(radians(A_ang))
b_len = k * sin(radians(B_ang))

xA, yA = 0.0, 0.0
xB, yB = c_len, 0.0
d = c_len
x_mid = (b_len**2 - a_len**2 + d**2) / (2 * d)
y_mid = np.sqrt(max(0.0, b_len**2 - x_mid**2))
xC = xA + x_mid
yC = y_mid

A = (xA, yA)
B = (xB, yB)
C = (xC, yC)

# Gráfico del triángulo
plt.figure(figsize=(6,6))
plt.fill([A[0], B[0], C[0]], [A[1], B[1], C[1]], alpha=0.4, edgecolor='black')
plt.plot([A[0], B[0], C[0], A[0]], [A[1], B[1], C[1], A[1]], color='black')
plt.text(A[0]-0.05, A[1]-0.05, f'A ({A_ang}°)', fontsize=12)
plt.text(B[0]+0.02, B[1]-0.05, f'B ({B_ang}°)', fontsize=12)
plt.text(C[0], C[1]+0.03, f'C ({C_ang}°)', fontsize=12)
plt.title('Triángulo con sus ángulos')
plt.gca().set_aspect('equal', adjustable='box')
plt.axis('off')
plt.show()

# Gráfico de barras con los ángulos
plt.figure(figsize=(6,4))
labels = ['Ángulo menor (m)', 'Otro', 'Tercero (m+4)']
plt.bar(labels, angulos)
for i,v in enumerate(angulos):
    plt.text(i, v+1, f'{v}°', ha='center', fontsize=11)
plt.ylim(0, max(angulos)+15)
plt.title('Medidas de los ángulos')
plt.ylabel('Grados')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

Problema nº 15.-
Una persona invierte en un producto una cantidad de dinero, obteniendo un 5% de beneficio. Por otra inversión en un segundo producto, obtiene un beneficio del 3,5%. Sabiendo que en total invirtió 10 000 €, y que los beneficios de la primera inversión superan en 300 € a los de la segunda, ¿cuánto dinero invirtió en cada producto?


```python
import matplotlib.pyplot as plt

# --- Cálculos ---
x = 650 / 0.085  # inversión 1 (5%)
y = 10000 - x    # inversión 2 (3.5%)

beneficio1 = 0.05 * x
beneficio2 = 0.035 * y

print(f"Inversión en el primer producto: {x:.2f} €")
print(f"Inversión en el segundo producto: {y:.2f} €")
print(f"Beneficio primer producto: {beneficio1:.2f} €")
print(f"Beneficio segundo producto: {beneficio2:.2f} €")

# --- Gráfico 1: Inversión en cada producto ---
plt.figure(figsize=(6,4))
labels_inv = ['Primer producto (5%)', 'Segundo producto (3.5%)']
valores_inv = [x, y]

plt.bar(labels_inv, valores_inv, color=['#4da6ff','#66ff99'])
for i,v in enumerate(valores_inv):
    plt.text(i, v+50, f'{v:.2f} €', ha='center', fontsize=11)
plt.title('Inversión en cada producto')
plt.ylabel('Euros')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

# --- Gráfico 2: Beneficio en cada producto ---
plt.figure(figsize=(6,4))
labels_ben = ['Beneficio 1 (5%)', 'Beneficio 2 (3.5%)']
valores_ben = [beneficio1, beneficio2]

plt.bar(labels_ben, valores_ben, color=['#ff9999','#99ccff'])
for i,v in enumerate(valores_ben):
    plt.text(i, v+10, f'{v:.2f} €', ha='center', fontsize=11)
plt.title('Beneficio obtenido en cada producto')
plt.ylabel('Euros')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
